# **Combine Chicago & New Orleans Data in collab using Python.**

# **Load the Data**

```python
import pandas as pd

# Load datasets

chicago = pd.read_csv('/content/Chicago_listings.csv.gz')
new_orleans = pd.read_csv('/content/new_orleans_listings (1).csv.gz')
```

```python
chicago.head(2)
```

```python
new_orleans.head(2)
```

# **Inspect Structure**

```python
print(chicago.shape)
print(new_orleans.shape)
```

```python
print(chicago.columns)
```

```python
print(new_orleans.columns)
```

# **Add City Column**

```python
chicago["city"] = "Chicago"
new_orleans["city"] = "New Orleans"
```

# **Clean Key Columns**

```python
# Convert price to numeric

for df in [chicago, new_orleans]:
    df["price"] = (
        df["price"]
        .replace('[\$,]', '', regex=True)
        .astype(float)
    )
```

```python
# Handle missing values
for df in [chicago, new_orleans]:
    df.drop_duplicates(inplace=True)
    df["reviews_per_month"].fillna(0, inplace=True)
```

# **Combine the Datasets**


```python
airbnb_combined = pd.concat(
    [chicago, new_orleans],
    axis=0,
    ignore_index=True
)
```

# **Validation**

```python
airbnb_combined["city"].value_counts()
airbnb_combined.info()
```

# **Export for Tableau**

```python
airbnb_combined.to_csv("airbnb_chicago_neworleans_combined.csv", index=False)
```
