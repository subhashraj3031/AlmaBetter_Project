# **Sales Performance Overview**
# Scenario 1: Monthly trend in orders shipped

Question addressed:
How do the number of orders shipped change over time across months?

**Justification:**
A line chart is the most appropriate visualization for this scenario because the data represents a time series with a natural chronological order (months). Line charts excel at showing trends, patterns, and direction of change over time. By connecting data points sequentially, the chart makes it easy to identify:

1. Overall growth or decline in orders

2. Seasonal fluctuations

3. Sudden increases or drops between months

In this context, the analytical objective is to understand sales momentum and seasonality, which a line chart communicates more intuitively than bars or tables.

# Scenario 2: Sales distribution across regions

Question addressed:
How are total sales distributed across different geographic regions?

Justification:
A bar chart is ideal for comparing magnitudes across discrete categories, such as regions. The horizontal orientation is especially effective here because:

1. There are many regions, and horizontal bars improve label readability

2. It allows for easy comparison of sales volumes across regions

3. Differences between high- and low-performing regions are immediately visible

This visualization aligns with the analytical objective of ranking and comparing regions, making it clear which regions dominate sales and which contribute less.

# Scenario 3: Sales by day of week and category

Question addressed:
How do sales vary by product category across different days of the week?

Justification:
A heatmap is the most effective choice when analyzing relationships between two categorical variables (day of week and product category) with a numerical measure (sales). It allows users to:

1. Quickly spot patterns and intensity levels through color variation

2. Identify high- and low-performing categoryday combinations

3. Compare multiple dimensions simultaneously without clutter

The analytical objective here is pattern detection rather than precise values, and a heatmap supports this by enabling fast visual scanning and comparison.

**Overall rationale**

Each chart type was selected to match:

1. The structure of the data (time-based, categorical, or multidimensional)

2. The key analytical task (trend analysis, comparison, or pattern recognition)

3. The need for clarity and interpretability for decision-making


Together, these visualizations provide a coherent and effective overview of sales performance from temporal, regional, and categorical perspectives.

# **Profitability Analysis**

# Scenario 1: Product categories with the highest total sales

Question addressed:
Which product categories generate the highest total sales?

Justification:
A bar chart is the most effective visualization for comparing aggregate values across discrete categories. Product categories (Office Supplies, Furniture, Technology) are nominal variables, and the objective is to compare total sales magnitude between them.

This chart:

1. Enables quick comparison of sales totals

2. Clearly highlights the top-performing category

3. Avoids unnecessary complexity for a simple ranking task

Because the goal is straightforward comparison rather than trend or composition analysis, a bar chart is the most clear and interpretable option.

# Scenario 2: Profit composition by subcategory and segment

Question addressed:
How is profit distributed across subcategories within each customer segment?

Justification:
A stacked bar chart is ideal when the analytical objective is to understand both:

1. Total profit per segment, and

2. How individual subcategories contribute to that total

This visualization:

* Shows overall profit differences between segments (Consumer, Corporate, Home Office)

* Reveals the relative contribution of each subcategory within a segment

* Allows comparison of composition patterns across segments

Using separate bars for segments while stacking subcategories balances detail and readability, making it suitable for multi-level profit analysis.

# Scenario 3: Discount vs quantity by segment

Question addressed:
What is the relationship between discount levels and quantity sold across customer segments?

Justification:
A scatter plot is the best choice for examining the relationship between two continuous variablesquantity and discountwhile also distinguishing categories (segments).

This chart:

1. Makes it easy to identify correlations or trends (e.g., higher quantity with higher discount)

2. Highlights outliers and unusual segment behavior

3. Supports comparison between segments using color or markers

Because the goal is relationship analysis rather than aggregation, a scatter plot provides the most analytical value.

# Scenario 4: Profit comparison by product and subcategory

Question addressed:
How does profit vary across products and subcategories, including negative profits?

Justification:
A treemap is particularly effective for visualizing hierarchical data with proportional values, such as product categories broken down into subcategories.

This visualization:

1. Displays relative profit contribution using area size

2. Shows hierarchical structure clearly

3. Allows quick identification of high-profit and loss-making subcategories

Efficiently handles a large number of categories without clutter

The treemap aligns with the analytical objective of understanding profit distribution and identifying problem areas within a complex product hierarchy.

**Overall visualization rationale**

Each chart type was selected based on:

* The nature of the data (categorical, continuous, hierarchical)

* The business question being asked

* The need for clarity, comparability, and insight extraction
