# **Project Summary -**

This capstone project focuses on predicting the monthly closing stock price of Yes Bank using machine learning techniques. Yes Bank, a prominent private sector bank in India, experienced significant volatility in its stock prices, particularly after 2018 due to the high-profile fraud case involving its former CEO. This situation created an opportunity to analyze historical stock price trends and build predictive models to understand whether machine learning algorithms can effectively forecast stock closing prices under such volatile conditions.

The dataset used for this project contains monthly stock price data from the bank’s inception, including Open, High, Low, and Close prices. The primary objective was to predict the closing price of the stock for a given month using the available numerical features. Since stock prices are continuous in nature, this problem was framed as a supervised regression task.

The project began with a comprehensive Exploratory Data Analysis (EDA) to understand the structure, distribution, and behavior of the dataset. The Date column was converted into a proper datetime format and set as the index to maintain chronological order. Visualization techniques using Matplotlib and Seaborn were employed to examine trends over time. The analysis revealed a steady growth trend in stock prices until 2018, followed by a sharp decline during the fraud crisis. Correlation analysis showed a strong positive relationship between the Open, High, Low, and Close prices, indicating that these variables are highly interdependent and suitable for predictive modeling.

Data preprocessing included checking for missing values, validating data types, and ensuring time-based splitting of the dataset to prevent data leakage. Since stock data is sequential, the training and testing sets were divided chronologically rather than randomly. The first 80% of the data was used for training and the remaining 20% for testing.

Multiple regression models were implemented, including Linear Regression and Random Forest Regressor. Linear Regression served as a baseline model to understand the linear relationship between features and the target variable. However, stock market behavior is often nonlinear and influenced by complex patterns. Therefore, the Random Forest model was introduced to capture nonlinear dependencies and improve prediction accuracy.

Model performance was evaluated using regression metrics such as Root Mean Squared Error (RMSE) and R² score. The Random Forest model outperformed Linear Regression, demonstrating better predictive capability and lower error values. Feature importance analysis further confirmed that High and Low prices played a significant role in predicting the closing price.

The final model provides valuable insights for stakeholders such as investors, financial analysts, and risk management teams. By forecasting closing prices, stakeholders can make informed investment decisions, assess potential risks, and identify abnormal market movements. Although stock price prediction is inherently uncertain due to external macroeconomic factors, this project demonstrates that machine learning models can effectively capture historical patterns and provide reasonably accurate forecasts.

In conclusion, this end-to-end machine learning project successfully implemented data preprocessing, exploratory analysis, feature engineering, model training, evaluation, and interpretation. The results highlight the effectiveness of ensemble learning techniques in financial time-series prediction and showcase how data-driven approaches can support strategic decision-making in the banking and investment sectors.
