```python
import pandas as pd

# Load datasets

chicago = pd.read_csv('/content/Chicago_listings.csv.gz')
new_orleans = pd.read_csv('/content/new_orleans_listings (1).csv.gz')
```

```python
print(new_orleans.columns)
```

```python
chicago["city"] = "Chicago"
new_orleans["city"] = "New Orleans"
```

```python
for df in [chicago, new_orleans]:
    df["price"] = (
        df["price"]
        .replace('[\$,]', '', regex=True)
        .astype(float)
    )
```

```python
for df in [chicago, new_orleans]:
    df.drop_duplicates(inplace=True)
    df["reviews_per_month"].fillna(0, inplace=True)
```

```python
airbnb_combined = pd.concat(
    [chicago, new_orleans],
    axis=0,
    ignore_index=True
)
```

```python
airbnb_combined["city"].value_counts()
airbnb_combined.info()
```

```python
airbnb_combined.to_csv("airbnb_chicago_neworleans_combined.csv", index=False)
```
