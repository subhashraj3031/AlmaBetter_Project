# **40. Type I and Type II errors**
# **Introductio**

40: What are Type I and Type II errors in hypothesis testing?

In hypothesis testing, decisions are made based on sample data. However, these decisions can sometimes be incorrect. These incorrect decisions are classified as Type I and Type II errors.

# Step-by-Step Explanation
**Step 1: Understand Hypotheses**

* Null hypothesis (H₀): No effect or no difference

* Alternative hypothesis (H₁): There is an effect or difference

Example:
* H₀: Drug has no effect
* H₁: Drug has an effect

**Step 2: Type I Error (False Positive)**

Definition:Rejecting the null hypothesis when it is actually true.

Example:
* You conclude the drug works, but it actually does not.

* Probability of Type I error = α (significance level, usually 0.05)

**Step 3: Type II Error (False Negative)**

Definition:Failing to reject the null hypothesis when it is false.

Example:
* You conclude the drug does not work, but it actually works.
* Probability of Type II error = β

Step 4: Summary Table
Error Type	Meaning	Example
* Type I - False positive	- Detect effect when none exists
* Type II - False negative	- Miss effect when it exists

# **Conclusion**

Type I error is rejecting a true null hypothesis, and Type II error is failing to reject a false null hypothesis. Both errors affect the reliability of statistical conclusions.
